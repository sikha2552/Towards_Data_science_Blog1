```python
from skimage.io import imread
im=imread('/content/drive/MyDrive/MY_Learnings/BLOGS/dog.jpg')
plt.imshow(im,cmap='gray')
plt.title('Image loaded from the local drive')
plt.axis('off')
plt.show()
```

```python
from skimage.io import imread
im_gray=imread('/content/drive/MyDrive/MY_Learnings/BLOGS/dog.jpg',as_gray=True)
plt.imshow(im_gray,cmap='gray')
plt.title('Image loaded from the local drive as grayscale')
plt.axis('off')
plt.show()
```

```python
print('The size of the color image is: {}'.format(im.shape))
print('The size of the grayscale image is: {}'.format(im_gray.shape))
```

```python
from skimage.transform import resize
new_H=100
new_W=100
im_resize=resize(im,[new_H,new_W])
fig,ax=plt.subplots(1,2)
ax[0].imshow(im)
ax[0].set_title('Original image')
ax[1].imshow(im_resize)
ax[1].set_title('Resized image')
for a in ax:
  a.axis('off')

print('The size of the  image is: {}'.format(im_resize.shape))
```

```python
#plot images
plt.subplot(121), imshow(im)
plt.title('Original Image')
plt.subplot(122), imshow(im_resize)
plt.title('Resized Image')
plt.show()
```

Play with the color channels

```python
from skimage.color import rgb2gray,rgb2hsv,rgb2yuv,rgb2ycbcr
im_gray=rgb2gray(im)
plt.subplot(121), imshow(im)
plt.axis('off')
plt.title('Original Image (RGB)')
plt.subplot(122), imshow(im_gray)
plt.axis('off')
plt.title('Grayscale Format')
plt.show()
```

```python
plt.subplot(121), imshow(im)
plt.axis('off')
plt.title('Original Image (RGB)')
im_hsv=rgb2hsv(im)
plt.subplot(122), imshow(im_hsv)
plt.axis('off')
plt.title('HSV format')
plt.show()
```

```python
plt.subplot(121), imshow(im)
plt.axis('off')
plt.title('Original Image (RGB)')
im_yuv=rgb2yuv(im)
plt.subplot(122), imshow(im_yuv)
plt.axis('off')
plt.title('YUV format')
plt.show()
```

```python
from skimage.color import rgb2lab
plt.subplot(121), imshow(im)
plt.axis('off')
plt.title('Original Image (RGB)')
im_lab=rgb2lab(im)
plt.subplot(122), imshow(im_lab)
plt.axis('off')
plt.title('CIElab format')
plt.show()
```
