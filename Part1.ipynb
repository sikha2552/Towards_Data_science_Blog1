Read and display images using skimage

```python
#import the required packages 
from skimage.io import imshow
from skimage import data
#import any preloaded image from data() module
image= data.chelsea()
imshow(image)
```

Read and display images using mtplotlib

```python
import matplotlib.pyplot as plt
from skimage import data
#import any preloaded image from data() module
image= data.chelsea()
plt.imshow(image)
plt.title('Preloaded image read from data module ')
plt.axis('off')
plt.show()
```

```python
from skimage.io import imread
im=imread('/content/drive/MyDrive/MY_Learnings/BLOGS/dog.jpg')
plt.imshow(im,cmap='gray')
plt.title('Image loaded from the local drive')
plt.axis('off')
plt.show()
```

```python
from skimage.io import imread
im=imread('/content/drive/MyDrive/MY_Learnings/BLOGS/dog.jpg',as_gray=True)
plt.imshow(im,cmap='gray')
plt.title('Image loaded from the local drive as grayscale')
plt.axis('off')
plt.show()
```
